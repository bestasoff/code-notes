## Дисбаланс классов

### Что делать при дисбалансе?
Нужно уточнить следующие моменты:
   * __надо обязательно уточнить природу задачи__: в чём причина дисбаланса, сколько классов, насколько серьёзный дисбаланс, какими данными и мета-данными мы располагаем;
   * __надо уточнить функцию ошибки (функционал качества)__. $F_1$-мера может расти не из-за того, что прогноз становится «адекватнее»;
   * __нужно уточнить контекст вопроса__;
   * __чтобы понять, как учитывать дисбаланс необходимо понять, насколько хорошо наши модели в данной задаче могут быть откалиброваны, какая геометрия данных и распределения классов__;
   
## Перебалансировка данных

### Виды перебалансировки
   * недосемлировние (_undersampling_);
   * пересемплировние (_oversampling_);

## Undersampling

### NearMiss
#### NearMiss-1
NearMiss-1 – из большего класса выбираем объекты, у которых среднее расстояний до N __ближайших__ малого класса наименьшее, т.е. из граничной зоны.
#### NearMiss-2
NearMiss-2 – из большего класса выбираем объекты, у которых среднее расстояний до N __дальних__ малого класса наименьшее.

### Tomek links
Tomek links – удалить объекты большого класса, образующие связи Томека (объекты двух разных классов образуют связь Томека, если нет объекта, который ближе к одному из них при этом являясь объектом другого класса).

### ENN
ENN – сделать скользящий контроль, например по 10 фолдам, удалить объекты большого класса, на которых ближайший сосед ошибается.

## Oversampling
### SMOTE (synthetic minority oversampling techniques)
Идея метода SMOTE: увеличить малый класс за счёт представителей выпуклых комбинаций пар. В методе SMOTE для точки малого класса выбирается один из k ближайших соседей и на отрезке между ними случайно выбирается новый объект. Почему-то нигде в описаниях не указывается, но в качестве k ближайших соседей рассматриваются только объекты малого класса.

### ADASYN
Метод ADASYN аналогичен SMOTE, но число объектов, которые генерируются с помощью объекта малого класса, пропорционально числу чужаков (объектов большого класса) в его окрестности.
